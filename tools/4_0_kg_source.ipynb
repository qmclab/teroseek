```python
import pandas as pd
from tqdm import tqdm

df_kg = pd.read_csv('.data/TA_KG_info.csv')
df_kg
```

```python
# 总结生源：
family_count = {}
genus_count = {}
species_count = {}
all_source = []
all_nodes = []
for _, row in tqdm(df_kg.iterrows()):
    T_id = row["T_id"]
    source_list = eval(row["source"])
    for source in source_list:
        _family = source.get("family", "")
        if _family:
            if _family not in family_count:
                family_count[_family] = 0
            family_count[_family] += 1
        _genus = source.get("genus", "")
        if _genus:
            if type(_genus) != list:
                _genus = [_genus]
                for __genus in _genus:
                    if __genus not in genus_count:
                        genus_count[__genus] = 0
                    genus_count[__genus] += 1
                    _species = source.get("species", "")
                    if _species:
                        if type(_species) != list:
                            _species = [_species]
                            for __species in _species:
                                if __species not in species_count:
                                    species_count[__species] = 0
                                species_count[__species] += 1
                                all_source.append({
                                    "family": _family,
                                    "genus": __genus,
                                    "species": __species
                                })
                                all_nodes.append({
                                    "T_id": T_id,
                                    "type": "kg_S",
                                    "family": _family,
                                    "genus": __genus,
                                    "species": __species
                                })
family_count = dict(sorted(family_count.items(), key=lambda item: item[1], reverse=True))
genus_count = dict(sorted(genus_count.items(), key=lambda item: item[1], reverse=True))
species_count = dict(sorted(species_count.items(), key=lambda item: item[1], reverse=True))
```

```python
df_source = pd.read_csv(".data/TA_KG_source_nodes.csv")
df_source
```

```python
# 根据 family 获取 T_id 列表
def family_T_id(family):
    df_tmp = df_source[df_source["family"] == family].copy()
    df_tmp.drop_duplicates(subset=["T_id"], inplace=True)
    return list(df_tmp["T_id"].values)

all_family = list(family_count.keys())
all_family = [a for a in all_family if a != "None"]
family_data = []
for _family in all_family:
    T_id_list = family_T_id(_family)
    if len(T_id_list) < 2:
        continue
    family_data.append({
        "family": _family,
        "_len": len(T_id_list),
        "T_id_list": T_id_list
    })
df_source_family = pd.DataFrame(family_data)
df_source_family
```

```python
# 根据 genus 获取 T_id 列表
def genus_T_id(genus):
    df_tmp = df_source[df_source["genus"] == genus].copy()
    df_tmp.drop_duplicates(subset=["T_id"], inplace=True)
    return list(df_tmp["T_id"].values)
all_genus = list(genus_count.keys())
all_genus = [a for a in all_genus if a != "None"]
genus_data = []
for _genus in all_genus:
    T_id_list = genus_T_id(_genus)
    if len(T_id_list) < 2:
        continue
    genus_data.append({
        "genus": _genus,
        "_len": len(T_id_list),
        "T_id_list": T_id_list
    })
df_source_genus = pd.DataFrame(genus_data)
df_source_genus
```

```python
# 根据 species 获取 T_id 列表
def species_T_id(species):
    df_tmp = df_source[df_source["species"] == species].copy()
    df_tmp.drop_duplicates(subset=["T_id"], inplace=True)
    return list(df_tmp["T_id"].values)
all_species = list(species_count.keys())
all_species = [a for a in all_species if a != "None"]
species_data = []
for _species in all_species:
    T_id_list = species_T_id(_species)
    if len(T_id_list) < 2:
        continue
    species_data.append({
        "species": _species,
        "_len": len(T_id_list),
        "T_id_list": T_id_list
    })
df_source_species = pd.DataFrame(species_data)
df_source_species
```

```python
print("Family count:", sum(df_source_family["_len"]))
df_source_family.to_csv(".data_KG/TA_len_family.csv", index=False)
print("Genus count:", sum(df_source_genus["_len"]))
df_source_genus.to_csv(".data_KG/TA_len_genus.csv", index=False)
print("Species count:", sum(df_source_species["_len"]))
df_source_species.to_csv(".data_KG/TA_len_species.csv", index=False)
```

```python
# sangji figure
def family_sangji_data(family):
    sangji_data = {}
    links = []
    df_tmp = df_source[df_source["family"] == family].copy()
    for _, row in df_tmp.iterrows():
        family = row["family"]
        genus = row["genus"]
        species = row["species"]
        edge_1 = f"{family}___{genus}"
        edge_2 = f"{genus}___{species}"
        if edge_1 not in sangji_data:
            sangji_data[edge_1] = 0
        if edge_2 not in sangji_data:
            sangji_data[edge_2] = 0
        sangji_data[edge_1] += 1
        sangji_data[edge_2] += 1

    for edge, count in sangji_data.items():
        if int(count) < 50:
            continue
        node_1, node_2 = edge.split("___")
        links.append({
            "source": node_1,
            "target": node_2,
            "value": count
        })
    nodes = list(set([link['source'] for link in links] + [link['target'] for link in links]))
    nodes_data = [{"name": node} for node in nodes]

    return nodes_data, links


family_sangji_data("Lamiaceae")

```
