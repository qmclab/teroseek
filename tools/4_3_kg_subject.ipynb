```python
import pandas as pd
from tqdm import tqdm

df_kg = pd.read_csv('.data/TA_KG_info.csv')
df_kg
```

```python
# 总结学科
subject_nodes = []
for _, row in df_kg.iterrows():
    T_id = row['T_id']
    subjects = eval(row['subject'])
    for subject in subjects:
        subject_nodes.append({
            'T_id': T_id,
            'type': 'kg_S',
            'subject': subject
        })        
```

```python
subject_nodes_df = pd.DataFrame(subject_nodes)
subject_nodes_df.to_csv(".data/TA_KG_subject_nodes.csv", index=False)
```
