```python
df = pd.read_csv("introduce_split.csv")[["custom_id", "source", "terpenoid", "keywords"]]
df
```

```python
source_dict = {}

for source_list in df["source"].tolist():
    for source in source_list.split(","):
        source = source.strip().replace("'","").replace("[","").replace("]","")
        if source not in source_dict:
            source_dict[source] = 0
        source_dict[source] += 1

sorted_dict = dict(sorted(source_dict.items(), key=lambda item: item[1], reverse=True))
sorted_dict
```

```python
compound_dict = {}

for compound_list in df["terpenoid"].tolist():
    for compound in compound_list.split(", "):
        compound = compound.strip().replace("'","").replace("[","").replace("]","")
        if compound not in compound_dict:
            compound_dict[compound] = 0
        compound_dict[compound] += 1

sorted_dict = dict(sorted(compound_dict.items(), key=lambda item: item[1], reverse=True))
sorted_dict
```
