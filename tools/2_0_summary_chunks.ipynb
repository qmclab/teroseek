```python
import pandas as pd
import os

parquet_list = [f"article/{a}" for a in os.listdir("article") if a.startswith("chunks")]
parquet_list

df = pd.concat(
    [pd.read_parquet(a) for a in parquet_list],
)
df
```

```python
df[["chunk_id", "summary","original_text"]].to_csv("query_extraction.csv", index=False, encoding="utf-8-sig")
```
