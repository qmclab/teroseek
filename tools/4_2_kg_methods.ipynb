```python
import json
import pandas as pd
data = []
with open("methods2.jsonl", "r") as f:
    for line in f:
        _data = json.loads(line)
        custom_id = _data["custom_id"]
        content = _data["response"]["body"]["choices"][0]["message"]["content"]
        content = content.replace("\n", "").replace("json", "").replace("`", "")
        try:
            data.append({
                "T_id": custom_id,
                "method": eval(content)
            })
        except:
            print(custom_id)
df_method = pd.DataFrame(data)
```

```python
df_method
```

```python
# 总结方法
df_kg = df_method
method_nodes = []
for _, row in df_kg.iterrows():
    T_id = row['T_id']
    methods = list(row['method'])
    for method_data in methods:
        method = method_data.get("method", "")
        metrial = method_data.get("metrial", "")
        result = method_data.get("result", "")
        method_nodes.append({
            'T_id': T_id,
            'type': "kg_M",
            'method': method,
            'metrial': metrial,
            'result': result
        })
method_nodes_df = pd.DataFrame(method_nodes)
method_nodes_df
```

```python
method_nodes_df.to_csv('.data/TA_KG_method_nodes.csv', index=False, encoding='utf-8-sig')
```
